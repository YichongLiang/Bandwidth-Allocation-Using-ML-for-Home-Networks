```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("20sec.csv")
```

```python
def filter_signal(df):
    filtered_df = df[df['Source'] == 'ba:5e:71:49:9b:51']
    if not pd.api.types.is_string_dtype(df['Signal strength (dBm)']):
        df['Signal strength (dBm)'] = df['Signal strength (dBm)'].astype(str)
    df['Signal strength (dBm)'] =  df['Signal strength (dBm)'].str.replace(r'[^-0-9]', '', regex=True).astype(int)
    signal_strength_list = []
    signal_strength_list.extend(filtered_df['Signal strength (dBm)'])
    return signal_strength_list

wifiSignal = filter_signal(df)
distance = np.linspace(0, 400, len(wifiSignal), endpoint=True)

plt.scatter(distance, wifiSignal)
plt.xlabel('Distance (cm)')
plt.ylabel('Signal Strength (dBm)')
plt.title('Signal Strength vs Distance (LOS Con)')
plt.show()
```
